```python
fig, ax = plt.subplots()

children = [10,20,5,11,24,100,90,51,56,80]

colors = ['red','green','yellow','orange','pink','marron','chilli','brown','light','tranparent color']

bar_labels = ['red','green','yellow','orange','pink','marron','chilli','brown','light','preet']

ax.bar(colors,children, color='red', label=bar_labels)


ax.set_ylabel('no of students')
ax.set_xlabel('color chossen')

ax.set_title('Fruit supply by kind and color')

plt.xticks(rotation='vertical')
```

```python
ps = pd.read_csv('MatplotlibDataSets/batsman_season_record.csv')

ps

# print(ps.shape[0])
```

```python
plt.bar(np.arange(ps.shape[0]) - 0.2 ,ps['2015'], width=0.2)
plt.bar(np.arange(ps.shape[0]),ps['2016'],width=0.2)
plt.bar(np.arange(ps.shape[0]) + 0.2,ps['2017'], width=0.2)

plt.xticks(np.arange(ps.shape[0]),ps['batsman'])

# IF WE CAN PLOT THIS TYPE OF INDEPENDENT GRAPH WE CAN USE ABOVE TRICK
```

```python
plt.bar(ps['batsman'],ps['2015'],label='2015')
plt.bar(ps['batsman'],ps['2016'],bottom=ps['2015'], label='2016')
plt.bar(ps['batsman'],ps['2017'],bottom=ps['2016'] + ps['2015'],label='2017')


plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Runs Scored by Batsmen in 2015, 2016, and 2017')

plt.legend()
plt.show()
```
